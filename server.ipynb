```python
import pandas as pd
import numpy as np
import json
df = pd.read_excel(r"C:\Users\Aman\Downloads\Data1.xlsx")
column_names = ["Order No.","No. of passengers", "Luggage(Kg)", "Distance(Km)", "Traffic(mins)", "Terrain(Km)", "Estimated Fare(₹)"]
# Create the dictionary with the column names as keys
order_passenger = {column: [] for column in column_names}
for index, row in df.iterrows(): #for eg. No. of passenger is key and values would be the value of the key (no. of pass)
    for column in column_names:
        order_passenger[column].append(row[column])
# print(order_passenger["No. of passengers"][1])
# print(order_passenger["No. of passengers"][3])
# print(order_passenger["Luggage(Kg)"][1])



order_passengers = {"No. of passengers": 4,
                   "Luggage(Kg)": 5,  # The luggage carried by passengers
                   "Distance(Km)": 50,
                   "Traffic(mins)": 20,  # Traffic is accommodated as waiting time
                   "Terrain(Km)": 12}  # Terrain is accommodated as Kms of underdeveloped roads
order_cargo = {"payload": 0,
               "Distance(Km)": 70,
               "Traffic(mins)": 20,
               "Terrain(Km)": 12}
n = len(order_passenger["No. of passengers"]) #Number of orders
x_ij = []
for row in range(6):
  order_n = []
  for coloumn in range(n):
    order_n.append(0)
  x_ij.append(order_n)

# print(x_ij)


# cost = x*distance + y*traffic + z*terrain



# constraint equations for our final multiple order value based input data
for i in range(n):

  if order_passenger["No. of passengers"][i]<=4 and order_passenger["Luggage(Kg)"][i]<=5:
    x_ij[0][i] = 1
    x_ij[3][i] = 1


  elif order_passenger["No. of passengers"][i]<=4 and order_passenger["Luggage(Kg)"][i]>5 and order_passenger["Luggage(Kg)"][i] <= 250:
    x_ij[1][i] = 1
    x_ij[4][i] = 1

  elif order_passenger["No. of passengers"][i]>4 and order_passenger["No. of passengers"][i]<=7:
    x_ij[2][i] = 1
    x_ij[5][i] = 1

# print(x_ij)



x = [16.5, 19.1295, 23, 10, 14, 18]  # cost per km for each type of vehicle
capital_cost = [4.2, 6.74, 8.1, 11.61, 12.49, 17.55]  # capital cost for each type of vehicle in lakhs
# print(x_ij)

y= [0.6, 0.8, 1.35, 1.5, 2.1, 2.5]
# print(x_ij)



#def matrix_multiplication(m1,m2):
 # result = []
  #for row in range(6):
   # order_n = []
 # for coloumn in range(n):
  #  order_n.append(0)
   # result.append(order_n)
  #for i in range(n):
   # for j in range(6):
    #  result[j][i] = m1[j] * m2[j][i]
 # return result

def matrix_multiplication(m1, m2):
  result = [[0 for _ in range(len(m2[0]))] for _ in range(len(m1))]
  for i in range(n):
    for j in range(6):
      result[j][i] = m1[j] * m2[j][i]
  return result

def matrix_multiplication2(m1,m2):
  result = [[0 for _ in range(len(m2[0]))] for _ in range(6)]
  for i in range(n):
    for j in range(6):
      result[j][i] = m1[i] * m2[j][i]
  return result



# print(x_ij)
# print(capital_cost)
cost_capital_final = matrix_multiplication(capital_cost,x_ij)
# print(cost_capital_final)

distance = []
for i in range(n):
  distance.append(order_passenger["Distance(Km)"][i])
# print(distance)

distance_int = matrix_multiplication(x,x_ij)
# print(x)
# print(distance_int)
cost_distance_final = matrix_multiplication2(distance, distance_int)
# print(cost_distance_final)



traffic_time = []
for i in range(n):
  traffic_time.append(order_passenger["Traffic(mins)"][i])
# print(traffic_time)

traffic_int = matrix_multiplication(y,x_ij)
cost_traffic_final = matrix_multiplication2(traffic_time, traffic_int)
# print(cost_traffic_final)



z = (16.5, 19.1295, 21, 12.0, 14.0, 16)
terrain = []
for i in range(n):
  terrain.append(order_passenger["Terrain(Km)"][i])
# print(terrain)

terrain_int = matrix_multiplication(z,x_ij)
# print(terrain_int)
cost_terrain_final = matrix_multiplication2(terrain, terrain_int)
# print(cost_terrain_final)



# print(x_ij)
# print(capital_cost)
cost_capital_final = matrix_multiplication(capital_cost,x_ij)
# print(cost_capital_final)

def matrix_sum(m1,m2,m3):
  sum_matrix = []
  for i in range(6):
    order_n = []
    for j in range(n):
      order_n.append(0)
    sum_matrix.append(order_n)
  for i in range(6):
    for j in range(n):
      sum_matrix[i][j] = m1[i][j] + m2[i][j] + m3[i][j]
  return sum_matrix


cost = matrix_sum(cost_distance_final, cost_traffic_final, cost_terrain_final)
cost_final = np.round(cost, decimals=0)
# print(cost_final)

Est_Fare = []
for i in range(n):
  Est_Fare.append(order_passenger["Estimated Fare(₹)"][i])
# print(Est_Fare)

profits = []
for row in range(6):
  order_n = []
  for coloumn in range(n):
    order_n.append(0)
  profits.append(order_n)

for i in range(6):
  a = 0
  for j in range(n):
    if cost_final[i][j] != 0:
      profits[i][j] = Est_Fare[a] - cost_final[i][j]
    a = a + 1
# print(profits)

orders = []
b = 0
car="abc"
for i in range(6):
  for j in range(n):
    if profits[i][j] != 0:
      [4.2, 6.74, 8.1, 11.61, 12.49, 17.55]
      if cost_capital_final[i][j] == 4.2:
        car = "mini CNG"
      elif cost_capital_final[i][j] == 6.74:
        car = "sedan CNG"
      elif cost_capital_final[i][j] == 8.1:
        car = "SUV CNG"
      elif cost_capital_final[i][j] == 11.61:
        car = "mini EV"
      elif cost_capital_final[i][j] == 12.49:
        car = "sedan EV"
      elif cost_capital_final[i][j] == 17.55:
        car = "SUV EV"
      orders.append((b,cost_capital_final[i][j],profits[i][j],car))
      b = b + 1

'''print(orders)
print(len(orders))'''



from pyomo.environ import *
from pyomo.opt import SolverFactory

def create_model(orders, capacity):
    model = ConcreteModel()

    # Sets
    model.orders = Set(initialize=range(1, len(orders) + 1))
    model.capacity = Param(initialize=capacity)

    # Parameters
    model.profits = Param(model.orders, initialize={i: order[2] for i, order in enumerate(orders, start=1)})
    model.weights = Param(model.orders, initialize={i: order[1] for i, order in enumerate(orders, start=1)})

    # Variables
    model.x = Var(model.orders, domain=Binary)

    # Objective
    def obj_rule(model):
        return sum(model.profits[i] * model.x[i] for i in model.orders)
    model.obj = Objective(rule=obj_rule, sense=maximize)

    # Capacity constraint
    def capacity_constraint_rule(model):
        return sum(model.weights[i] * model.x[i] for i in model.orders) <= model.capacity
    model.capacity_constraint = Constraint(rule=capacity_constraint_rule)

    # Additional constraint
    n = len(orders) // 2
    def additional_constraint_rule(model, i):
        return model.x[i] + model.x[i + n] <= 1 if i + n <= len(orders) else Constraint.Skip
    model.additional_constraint = Constraint(range(1, n + 1), rule=additional_constraint_rule)

    return model

def solve_model(orders, capacity):
    model = create_model(orders, capacity)
    solver = SolverFactory('glpk')
    solver.solve(model)

    selected_orders = [i for i in model.orders if model.x[i].value == 1]
    max_profit = model.obj()
    selected_cars = [i for i in model.orders]

    return max_profit, selected_orders

orders
capacity = 20
max_profit, selected_orders = solve_model(orders, capacity)
print("Maximum Profit:", max_profit)
print("Selected Orders:", selected_orders)
print("Selected Cars:",)
output = "abc"
Mini_CNG = 0
Sedan_CNG = 0
SUV_CNG = 0
Mini_EV = 0
Sedan_EV = 0
SUV_EV = 0
for i in range(len(selected_orders)):
  output = orders[selected_orders[i]][3]
  print(output)
  if output == "mini CNG":
    Mini_CNG = Mini_CNG + 1
  if output == "sedan CNG":
    Sedan_CNG = Sedan_CNG + 1
  if output == "SUV CNG":
    SUV_CNG = SUV_CNG + 1
  if output == "mini EV":
    Mini_EV = Mini_EV + 1
  if output == "sedan EV":
    Sedan_EV = Sedan_EV + 1
  if output == "SUV EV":
    SUV_EV = SUV_EV + 1

print(Mini_CNG)
print(Sedan_CNG)
print(SUV_CNG)
print(Mini_EV)
print(Sedan_EV)
print(SUV_EV)

array = [0,0,0,0,0,0]
array[0] = Mini_CNG
array[1] = Sedan_CNG
array[2] = SUV_CNG
array[3] = Mini_EV
array[4] = Sedan_EV
array[5] = SUV_EV

array_json = json.dumps(array)

import socket
import pandas as pd

vehicles = {"CNG": {"Mini": Mini_CNG, "Sedan": Sedan_CNG, "SUV": SUV_CNG},
            "EV": {"Mini": Mini_EV, "Sedan": Sedan_EV, "SUV": SUV_EV}}

data = pd.read_excel(r"C:\Users\Aman\Downloads\Orders_current.xlsx")
column_names = ["Order No.","No. of passengers", "Luggage(Kg)", "Distance(Km)", "Traffic(mins)", "Terrain(Km)", "Estimated Fare(₹)"]
# Create the dictionary with the column names as keys
order_passenger = {column: [] for column in column_names}
for index, row in data.iterrows(): #for eg. No. of passenger is key and values would be the value of the key (no. of pass)
    for column in column_names:
        order_passenger[column].append(row[column])


server = socket.socket()
server.bind(('localhost', 9999))
server.listen(1000)
print('waiting for connection')

# Client-Server interaction module
while True:
    client, addr = server.accept()
    persona = client.recv(1024).decode()
    if persona == 'Customer':
        client.send(bytes(array_json, 'utf-8'))

    print('Connected with Firm Client:', addr)
    vehicle_type = client.recv(1024).decode()
    print(vehicle_type)
    if vehicle_type == 'C1':
        vehicles["CNG"]["Mini"] += 1
    if vehicle_type == 'C2':
        vehicles["CNG"]["Sedan"] += 1
    if vehicle_type == 'C3':
        vehicles["CNG"]["SUV"] += 1
    if vehicle_type == 'E1':
        vehicles["EV"]["Mini"] += 1
    if vehicle_type == 'E2':
        vehicles["EV"]["Sedan"] += 1
    if vehicle_type == 'E3':
        vehicles["EV"]["SUV"] += 1

    if vehicle_type == 'CNG Mini':
        vehicles["CNG"]["Mini"] -= 1
    if vehicle_type == 'CNG Sedan':
        vehicles["CNG"]["Sedan"] -= 1
    if vehicle_type == 'CNF SUV':
        vehicles["CNG"]["SUV"] -= 1
    if vehicle_type == 'EV Mini':
        vehicles["EV"]["Mini"] -= 1
    if vehicle_type == 'EV Sedan':
        vehicles["EV"]["Sedan"] -= 1
    if vehicle_type == 'EV SUV':
        vehicles["EV"]["SUV"] -= 1
    print(vehicles)
    client.close()
```
